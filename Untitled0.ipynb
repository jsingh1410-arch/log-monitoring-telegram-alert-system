```python
# Program demonstrating assignment operators
num = 5
# Assignment operations
print("Intial value: num =", num)
num +=3 # Equivalent to num = num + 3
print("After num += 3:", num)
num -+ 2 # Equivalent to num = num - 2
print("After num -+ 2:", num)
num *= 4 # Equivalent to num = num * 4
print("After num *= 4:", num)
num /= 2 # Equivalent to num = num / 2
print("After num /= 2:", num)
num %= 3 # Equivalent to num = num % 3
print("After num %= 3:", num)
```

```python
# Program demonstrating comparison operators
x = 10
y = 20
# Comparison operations
print("x == y:", x == y)
print("x != y:", x != y)
print("x > y:", x > y)
print("x < y:", x < y)
print("x >= y:", x >= y)
print("x <= y:", x <= y)
```
