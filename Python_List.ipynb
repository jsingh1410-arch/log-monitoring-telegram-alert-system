```python
# Slice from the start
b = "Hello, World!"
print(b[:5])
# Slice to the End
b = "Hello, World!"
print(b[2:])
# Negative Indexing
b = "Hello, World!"
print(b[-5:-2]) # -5:-2 = orl
```
